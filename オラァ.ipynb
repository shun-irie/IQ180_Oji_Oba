```python
import numpy as np
import matplotlib.pyplot as plt

IQs = np.random.normal(loc=100, scale=15, size=130000000)
count, Xs = np.histogram(IQs, range=(0, 200), bins=200)
plt.plot(Xs[1:], count, 'k-')
plt.axvline(x=180)

idx_180 = np.where(Xs >= 180)[0][0]
above_180 = np.sum(count[idx_180:])

plt.title(f'IQ >= 180: {above_180}')
plt.show()
```
